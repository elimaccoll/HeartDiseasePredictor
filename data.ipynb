```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/heart.csv")
df.head(5)
```

```python
# Ages
ageN = df.loc[df["HeartDisease"] == 0, "Age"]
ageY = df.loc[df["HeartDisease"] == 1, "Age"]

print(sorted(list(ageN)))
print(sorted(list(ageY)))
```

```python
# Cholesterol
cholN = df.loc[df["HeartDisease"] == 0, "Cholesterol"]
cholY = df.loc[df["HeartDisease"] == 1, "Cholesterol"]

print(sorted(list(cholN)))
print(sorted(list(cholY)))
```

```python
# Max HR
maxhrN = df.loc[df["HeartDisease"] == 0, "MaxHR"]
maxhrY = df.loc[df["HeartDisease"] == 1, "MaxHR"]

print(sorted(list(maxhrN)))
print(sorted(list(maxhrY)))
```

```python
# Sex
sexN = df.loc[df["HeartDisease"] == 0, "Sex"]
sexY = df.loc[df["HeartDisease"] == 1, "Sex"]

sexNCounts = sorted(list(sexN))
sexYCounts = sorted(list(sexY))


counts = {"N": {"M": 0, "F": 0}, "Y": {"M": 0, "F": 0}}
for x in sexNCounts:
    counts["N"][x] += 1
for x in sexYCounts:
    counts["Y"][x] += 1
print(counts)
```

```python
# Fasting BS
fastingN = df.loc[df["HeartDisease"] == 0, "FastingBS"]
fastingY = df.loc[df["HeartDisease"] == 1, "FastingBS"]

fastingNCounts = sorted(list(fastingN))
fastingYCounts = sorted(list(fastingY))

counts = {"N": {"Y": 0, "N": 0}, "Y": {"Y": 0, "N": 0}}
for x in fastingNCounts:
    label = "Y"
    if x == 0:
        label = "N"
    counts["N"][label] += 1
for x in fastingYCounts:
    label = "Y"
    if x == 0:
        label = "N"
    counts["Y"][label] += 1
print(counts)
```

```python
# Fasting BS
exerciseN = df.loc[df["HeartDisease"] == 0, "ExerciseAngina"]
exerciseY = df.loc[df["HeartDisease"] == 1, "ExerciseAngina"]

exerciseNCounts = sorted(list(exerciseN))
exerciseYCounts = sorted(list(exerciseY))

counts = {"N": {"Y": 0, "N": 0}, "Y": {"Y": 0, "N": 0}}
for x in exerciseNCounts:
    counts["N"][x] += 1
for x in exerciseYCounts:
    counts["Y"][x] += 1
print(counts)
```

```python
# Oldpeaks
oldpeakN = df.loc[df["HeartDisease"] == 0, "Oldpeak"]
oldpeakY = df.loc[df["HeartDisease"] == 1, "Oldpeak"]

print(sorted(list(oldpeakN)))
print(sorted(list(oldpeakY)))
```

```python
# Chest Pain Type
chestN = df.loc[df["HeartDisease"] == 0, "ChestPainType"]
chestY = df.loc[df["HeartDisease"] == 1, "ChestPainType"]

chestN = sorted(list(chestN))
chestY = sorted(list(chestY))

counts = {"N": {"ATA": 0, "NAP": 0, "ASY": 0, "TA": 0}, "Y": {"ATA": 0, "NAP": 0, "ASY": 0, "TA": 0}}
for x in chestN:
    counts["N"][x] += 1
for x in chestY:
    counts["Y"][x] += 1
print(counts)
```
